```python
class Solution:
    def isValid(self, s: str) -> bool:
        hash_map = {
            ')':'(', ']':'[', '}':'{'
        }
        stack=[]
        for char in s:
            if char in hash_map:
                if stack and hash_map[char]==stack[-1]:
                    stack.pop()
                else:
                    return False    
            else:
                stack.append(char)
        return len(stack)==0        
```
